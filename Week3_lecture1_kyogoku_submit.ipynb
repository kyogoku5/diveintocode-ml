## 【問題4】データ探索

### ▼欠損値を処理のために文字データを数値化

```python
ohe_df = pd.get_dummies(df)
ohe_df.head()
```

### ▼欠損箇所にその列の平均を代入

```python
from sklearn.preprocessing import Imputer
im = Imputer(missing_values="NaN",strategy='mean',axis=0)
im = im.fit(ohe_df)
df_imputed = im.transform(ohe_df)

#DataFrame型に変換
df2 = pd.DataFrame(df_imputed,columns=list(ohe_df.columns))
df2.head()
```

### ▼相関係数行列を作成

```python
df_corr = df2.corr()
df_corr.head()
```

### ▼相関係数行列をヒートマップで表示

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,20))
sns.heatmap(df_corr,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True, # True:格子の中に値を表示
            fmt='.1f',
            xticklabels=df_corr.columns.values,
            yticklabels=df_corr.columns.values
           )
plt.show()
```
