# ■第2週_授業前課題2_ダミーデータの作成

 ## 【問題1】乱数の作成

```python
import numpy as np

average = [-3,0]
procession = [[1.0, 0.8], [0.8, 1.0]]

np.random.seed(0)
random = np.random.multivariate_normal(average, procession ,500)

'''
■注意点
各行の左側がxの値、右側がyの値である
つまり横一列で１個の点の位置を表している
'''
random
```

## 【問題2】散布図による可視化

```python
import matplotlib.pyplot as plt
plt.title("scatter image")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(random[:,0], random[:,1])
plt.show()
```

## 【問題3】ヒストグラムによる可視化

```python
plt.hist(random[:,0],bins=50)
plt.title("histgram image1")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-5.5, 2.5)
plt.show()
```

```python
plt.hist(random[:,1],bins=50)
plt.xlim(-5.5, 2.5)
plt.title("histgram image2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
average2 = [0,-3]
procession2 = [[1.0, 0.8], [0.8, 1.0]]

np.random.seed(0)
random2 = np.random.multivariate_normal(average2, procession2 ,500)

plt.title("scatter image")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(random[:,0], random[:,1], label = '0')
plt.scatter(random2[:,0], random2[:,1], label = '1')

plt.legend()
plt.show()
```

## 【問題5】データの結合

```python
random_total = np.vstack((random, random2) )
random_total.shape
```
